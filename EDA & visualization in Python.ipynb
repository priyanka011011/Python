# Visualization in python

##### Importing the necessary packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.image import imread
from scipy import linalg as lg
```

##### Defining the image variable

```python
himym = imread("himym.jpg")
print(type(himym))
```

```python
print(himym)
```

##### Reshaping

```python
n = himym.size
N = 4
M = n//N
reshape_himym = himym.reshape((N,M))
print(reshape_himym) 
```

##### Transpose

```python
himym_trans = himym.transpose()
print(himym_trans)
```

```python
n = himym_trans.size
N = 4
M = n//N
reshape_himym_trans = himym_trans.reshape((N,M))
print(reshape_himym_trans)
```

##### Addition

```python
himym_add = reshape_himym + reshape_himym_trans
print(himym_add)
```

##### Subtraction

```python
himym_subtract = reshape_himym - reshape_himym_trans
print(himym_subtract)
```

##### Multiplication

```python
himym_multiply = reshape_himym * reshape_himym_trans
print(himym_multiply)
```

##### Square root

```python
np.sqrt(reshape_himym)
```

### Visualization

##### Line Plot

```python
data = pd.read_csv('himym_episodes.csv')
data
plt.figure(figsize = (12,8))

season_total = data[:22]["episode_num_in_season"]
votes_1 = data[:22]["us_viewers"]
votes_2 = data[22:44]["us_viewers"]
votes_3 = data[44:66]["us_viewers"]

plt.style.use('fivethirtyeight')
plt.plot(season_total, votes_1, linestyle='--', linewidth = 2, marker = '.',markeredgecolor = 'black', label='season 1 = us_viewers')
plt.plot(season_total, votes_2, linewidth = 2, marker = '.',markeredgecolor = 'black',label='season 2 = us_viewers')
plt.plot(season_total, votes_3, linewidth = 2, marker = '.', markeredgecolor = 'black' ,label='season 3 = us_viewers')

plt.xlabel('Episode Number', size = 12)
plt.ylabel('Votes', size = 12)
plt.title('himym - Votes', size = 18)
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

##### Bar Graph

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize = (12, 8))

season_total = data[:22]["episode_num_in_season"]
votes_4 = data[66:88]["us_viewers"]
votes_5 = data[88:110]["us_viewers"]
votes_6 = data[110:132]["us_viewers"] 
           
x_indexes = np.arange(len(season_total))
width = 0.25
           
plt.bar(x_indexes - width, votes_4, width = width, label='season 4 = Votes')
plt.bar(x_indexes, votes_5, width = width, label='season 5 = Votes')         
plt.bar(x_indexes + width, votes_6, color='#467200', width = width, label='season 6 = Votes')
plt.xticks(ticks = x_indexes, labels = season_total)
plt.xlabel('Episode Number')
plt.ylabel('Votes')
plt.title('himym - Votes')
plt.legend()
plt.tight_layout()
plt.show()
```

##### Stack Plot

```python
plt.figure(figsize = (12, 8))
plt.style.use("fivethirtyeight")

season_total = data[:22]["episode_num_in_season"]
votes_7 = data[138:160]["us_viewers"]
votes_8 = data[160:182]["us_viewers"]
votes_9 = data[182:204]["us_viewers"]

labels = ['votes_7', 'votes_8', 'votes_9']
colors = ['#6d904f', '#fc4f30', '#008fd5']

plt.stackplot(season_total, votes_7, votes_8, votes_9, labels=labels, colors=colors)
plt.legend(loc=(0.07, 0.05))
plt.title("himym - Vote")
plt.tight_layout()

plt.show()
```

##### Scatter Plot

```python
plt.style.use('fivethirtyeight')

plt.figure(figsize = (12, 8))

viewer_1 = data[:24]["us_viewers"]
viewer_2 = data[24:48]["us_viewers"]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5, 6, 9, 1, 3]
plt.scatter(viewer_1, viewer_2, s = 100, c = colors, 
            cmap = 'twilight_shifted', edgecolor = 'black', linewidth = 1, alpha = 0.75)

cbar = plt.colorbar()
cbar.set_label('number range')

plt.title('himym - us_viewers')
plt.xlabel('us_viewers - season 1')
plt.ylabel('us_viewers - season 2')
plt.tight_layout()

plt.show()
```
